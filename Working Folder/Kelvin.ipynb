```python
# Handle missing values - Drop columns with more than 40% missing values
threshold = 0.4 * data.shape[0]
data.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
# Fill remaining missing values with median for numerical and mode for categorical
def fill_missing_values(col):
    if data[col].dtype == 'object':
        data[col].fillna(data[col].mode()[0], inplace=True)
    else:
        data[col].fillna(data[col].median(), inplace=True)

# Apply the function to each column
for col in data.columns: # Iterate through columns
    fill_missing_values(col)
```
